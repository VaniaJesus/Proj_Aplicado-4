# **Introdução**
A poluição do ar é um dos maiores desafios ambientais enfrentados por grandes centros urbanos ao redor do mundo. O Objetivo de Desenvolvimento Sustentável (ODS) 11, que aborda a necessidade de tornar as cidades e os assentamentos humanos inclusivos, seguros, resilientes e sustentáveis, destaca a importância de melhorar a qualidade do ar e a saúde pública. Em cidades como São Paulo, com altos índices de poluentes atmosféricos devido ao tráfego intenso e à atividade industrial, a busca por soluções inovadoras para mitigar os efeitos da poluição é urgente.

Neste contexto, a tecnologia de tintas absorventes de poluentes emerge como uma alternativa promissora. Essa tinta é capaz de capturar e neutralizar gases poluentes, como óxidos de nitrogênio (NOx) e dióxido de carbono (CO2), ajudando a purificar o ar nas áreas onde é aplicada. O uso dessa tecnologia em grandes áreas externas de imóveis, como fachadas de edifícios e residências, tem o potencial de reduzir significativamente os níveis de poluição em zonas urbanas densamente povoadas.

O presente trabalho propõe uma política de incentivo à adoção dessa tecnologia em São Paulo, com a oferta de redução no Imposto Predial e Territorial Urbano (IPTU) para condomínios, residências e empresas que implementarem a tinta absorvente de poluentes em suas fachadas. Ao combinar a inovação tecnológica com políticas públicas, o objetivo é contribuir para a melhoria da qualidade do ar e, consequentemente, da saúde pública na cidade.



# **Motivações e justificativa**
